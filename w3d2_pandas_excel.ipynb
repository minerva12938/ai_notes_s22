```python
df1 = pd.DataFrame({'A' : range(4), 'B' : range(4,8), 'C' : range(9, 13), 'D' : range(13, 17)})
print(df1)
```

```python
df2 = pd.DataFrame({'A' : range(40, 44), 'B' : range(44, 48), 'C' : range(49, 53), 'D' : range(53, 57)})
print(df2)
```

```python
#axis == 0; concat range wise
df3 = pd.concat([df1, df2])

df3
```

```python
#duplicate rows bc there are 2 0 rows
df3.loc[0]
```
