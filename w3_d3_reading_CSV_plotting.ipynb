```python
import pandas as pd
import numpy as np

#read in all data
nfl_data = pd.read_csv("data.csv")
np.random.seed(0)
```

```python
nfl_data.head()
```

```python
missing_values_count = nfl_data.isnull().sum()
print(missing_values_count)
missing_values_count[0:10]
```

```python
#every line has na values so only an empty set is left
nfl_data.dropna()
```

```python
col_with_na = nfl_data.dropna(axis = 1)
col_with_na.head()

print(nfl_data.shape[1])
print(col_with_na.shape[1])
```

```python
subset_nfl = nfl_data.loc[:, 'EPA':'Season'].head()
subset_nfl
```

```python
subset_nfl.fillna(0)
```

```python
#fill with what's below it
subset_nfl.fillna(method = 'bfill', axis = 0)

#if there's nothing below it, fill as 0
subset_nfl.fillna(method = 'bfill', axis = 0).fillna(0)
```

```python
#can also use 'ffil' == forward fill which fills with what's above it
subset_nfl.fillna(method = 'ffill', axis = 0).fillna(0)
```
